```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
import os
from FCN_2D import FCN_2D
import matplotlib.pyplot as plt
```

```python
trains = [{'channels': 8, 'layers': 8},
          {'channels': 8, 'layers': 32},
          {'channels': 8, 'layers': 64},
          {'channels': 1, 'layers': 8},
          {'channels': 1, 'layers': 32},
          # {'channels': 1, 'layers': 64}
        ]
```

```python
def plot_info(df_record, title):
    plt.figure(figsize=(8, 5))
    plt.plot(df_record['epoch'], df_record['train_loss'], label='Train Loss')
    plt.plot(df_record['epoch'], df_record['valid_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
for train in trains:
    root_folder = os.getcwd()
    train_name = f'FCN_2D_{train["channels"]}ch_{train["layers"]}lyr'
    file_path = os.path.join(root_folder, 'results', train_name)
    csv_train = pd.read_csv(os.path.join(file_path, 'train_record.csv'))
    plot_info(csv_train, train_name)
    
    
```
